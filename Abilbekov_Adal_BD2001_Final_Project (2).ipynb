```python
# Firstly prepare our data by making new DF for easier plotting
# From the tables above found out top ranking universities by Times
cal_tech = df1[df1.university_name=='California Institute of Technology']
stanford = df1[df1.university_name=='Stanford University']
cambridge = df1[df1.university_name=='University of Cambridge']
oxford = df1[df1.university_name=='University of Oxford']
mit = df1[df1.university_name=='Massachusetts Institute of Technology']

# Changing the datatype from object to float
cal_tech.loc[:,'world_rank'] = cal_tech.loc[:,'world_rank'].astype(float)
stanford.loc[:,'world_rank'] = stanford.loc[:,'world_rank'].astype(float)
cambridge.loc[:,'world_rank'] = cambridge.loc[:,'world_rank'].astype(float)
oxford.loc[:,'world_rank'] = oxford.loc[:,'world_rank'].astype(float)
mit.loc[:,'world_rank'] = mit.loc[:,'world_rank'].astype(float)

# And here we are plotting the line plot for Times
plt.figure(figsize=(20, 10));

sns.set_style("darkgrid")
sns.lineplot(x='year', y='world_rank', data=stanford, label='Stanford');
sns.lineplot(x='year', y='world_rank', data=cal_tech, label='CalTech');
sns.lineplot(x='year', y='world_rank', data=cambridge, label='Cambridge');
sns.lineplot(x='year', y='world_rank', data=oxford, label='Oxford');
sns.lineplot(x='year', y='world_rank', data=mit, label='MIT');

plt.title('Times University Rankings, 2011-2016', fontsize=16);
plt.xlabel('Year', fontsize=16);
plt.ylabel('Rank', fontsize=16);
plt.grid(True)
plt.legend(loc='upper left', fontsize=16);
```

```python
# Making a slice of data for only 2015
rank2015 = df1[df1.year == 2015]
rank2015 = rank2015.head(100)
rank_serie = rank2015['world_rank'].str.replace("=", "")#using string method because some data consists other elemnts
rank2015.loc[:,'world_rank'] = rank_serie
rank2015['world_rank'] = rank2015['world_rank'].astype(float)
rank2015['income'] = rank2015['income'].astype(float)
rank2015.head()
```

```python
#Giving values for the X and Y for three scatter plots
x = rank2015['teaching']
y = rank2015['world_rank']

x1 = rank2015['research']
y1 = rank2015['world_rank']

x2 = rank2015['income']
y2 = rank2015['world_rank']

plt.figure(figsize = (28,20))

sns.set_style("darkgrid")
plt.subplot(3, 2, 1)
plt.scatter(x, y)
plt.title('World rank vs Teaching score(Times)')
plt.xlabel("World rank");
plt.ylabel("Teaching score");
m, b = np.polyfit(x, y, 1);#calculating the coefficentsfor linear regression
plt.plot(x, m*x+b, color='blue');#making the linear regression line

plt.subplot(3, 2, 3)
plt.scatter(x1, y1, color='yellow')
plt.title('World rank vs Research score(Times)')
plt.xlabel("World rank");
plt.ylabel("Research score");
m, b = np.polyfit(x1, y1, 1);
plt.plot(x, m*x+b, color='green');

plt.subplot(3, 2, 5)
plt.scatter(x2, y2, color='cornflowerblue')
plt.title('World rank vs Income score(Times)')
plt.xlabel("World rank");
plt.ylabel("Income score");
m, b = np.polyfit(x2, y2, 1);
plt.plot(x, m*x+b, color='hotpink');
```
